```python
import re
import json
import tweepy
import requests
from keys import consumer_key, consumer_secret, access_token, access_token_secret
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
class MyStreamListener(tweepy.Stream):

    def on_status(self, status):
#         if (not status.retweeted) and ('RT @' not in status.text):
        text = re.sub(r"http\S+", "", re.sub('\s+', ' ', status.text))
        cleaned_text = clean_string(text)

        print(cleaned_text)

        timestamp = status.created_at
        sentiment = sid.polarity_scores(cleaned_text)

        print(timestamp, sentiment)
        
        url = 'https://rn0dei6c4j.execute-api.us-east-2.amazonaws.com/canaal-near-stage/canaal-api'
        data = {'tweet': cleaned_text,
                'pos': round(float(sentiment['pos']),3),
                'neu': round(float(sentiment['neu']),3),
                'neg': round(float(sentiment['neg']),3)}

        response = requests.post(url, data = json.dumps(data))
            
def clean_string(tweet):
    return re.sub(r'[^a-zA-Z ]', '', tweet)
```

```python
myStreamListener = MyStreamListener(consumer_key, consumer_secret, access_token, access_token_secret)
myStreamListener.filter(track=['#NEAR','$NEAR','#near','$near','near protocol','near nft'])
```
